# job title simplefier

```python
def title(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
            return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    if 'manager' in title.lower():
        return 'manager'
    if 'director' in title.lower():
        return 'director'
    else:
        return 'na'
def seniority(title):
    if 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower() or 'sr.' in title.lower() or 'sr' in title.lower():
        return 'sr'
    elif 'junior' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
df['job_simp'] = df['Job Title'].apply(title)
df.job_simp.value_counts()
```

```python
df['senority'] = df['Job Title'].apply(seniority)
df.senority.value_counts()
```

# fix states

```python
df.job_state.value_counts()
```

```python
df['job_state']=df.job_state.apply(lambda x : 'CA' if x.strip().lower()=='california' or x.strip().lower()=='united states' or x.strip().lower()=='remote'  or x.strip().lower()=='utah' or x.strip().lower()=='new jersey' or x.strip().lower()=='texas' else x.strip())
df.job_state.value_counts()
```

# job description length

```python
df['desc_len'] = df['Job Description'].apply(lambda x : len(x))
df['desc_len']
```

# competetor length

```python
df['no_of_Competitors']=df.Competitors.apply(lambda x: len(x.split(',')) if x!='-1' else 0)
df['no_of_Competitors']
```

# hourly wages amount

```python
df['min_salary'] = df.apply(lambda x: x.min_salary*2 if x.hourly ==1 else x.min_salary, axis =1)
df['max_salary'] = df.apply(lambda x: x.max_salary*2 if x.hourly ==1 else x.max_salary, axis =1)
```
